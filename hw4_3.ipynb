```python
import math
```

```python
def gsfunc(mu, sigmaf, x):
    f1 = 1 / (math.pow(2 * math.pi, 0.5) * math.pow(sigmaf, 0.5))
    f2 = math.exp(-1 * math.pow(x - mu, 2) / (2 * sigmaf))
    ret = f1 * max(f2, 0.001)
    return ret
```

```python
class gs():
    def run(self, data, loopcnt=100, K=2, eps = 0.00001):
        self.K = K
        self.eps = eps
        datanum = len(data)
        alpha = [1 / float(K)] * K  # alpha初始值为1/K
        sigmaf = [1.0] * K  # sigmaf初始值为1
        sigma = [1.0] * K
        mu = [0.0] * K  # mu初始值为0
        print("alpha", alpha)
        print('sigmaf', sigmaf)
        print('mu', mu)

        def gamma(j, k):
            return alpha[k] * gsfunc(mu[k], sigmaf[k], data[j]) \
                   / sum([alpha[n] * gsfunc(mu[n], sigmaf[n], data[j]) \
                          for n in range(self.K)])

        for i in range(loopcnt):
            print('loopcnt:', i)
            alpha_old = alpha.copy()
            #print("alpha_old", alpha_old)
            mu_old = mu.copy()
            sigmaf_old = sigmaf.copy()



            for k in range(self.K):
                sum1 = sum([gamma(j, k) for j in range(datanum)])
                # print(sum1)
                sigmaf[k] = sum([gamma(j, k) * math.pow(data[j] - mu[k], 2)
                                 for j in range(datanum)]) \
                            / sum1
                mu[k] = sum([gamma(j, k) * data[j] for j in range(datanum)]) / sum1
                alpha[k] = sum1 / datanum
            print("alpha", alpha)

            # print('sigmaf', sigmaf)
            sigma = [math.pow(s, 0.5) for s in sigmaf]
            print('sigma', sigma)

            # print('sigma', list(map(lambda sigma: math.pow(sigma, 0.5), sigmaf)))

            print('mu', mu)

            alpha_error = [alpha_old[p] - alpha[p] for p in range(self.K)]
            sigmaf_error = [sigmaf_old[p] - sigmaf[p] for p in range(self.K)]
            mu_error = [mu_old[p] - mu[p] for p in range(self.K)]
            errors = alpha_error + sigmaf_error + mu_error
            print("error", errors)

            num = 0
            for error in errors:
                if error == 0:
                    num += 1
            if num == len(errors):
                print(i)
                break
```

```python
if __name__ == '__main__':
    data = [-67, -48, 6, 8, 14, 16, 23, 24, 28, 29, 41, 49, 56, 60, 75]
    gsins = gs()
    gsins.run(data)
```
